### Работа с пропусками

```python
raw_data.isna().sum()
```

```python
pd.DataFrame(round(raw_data.isna().mean()*100,)).style.background_gradient('OrRd')
```

```python
raw_data.loc[raw_data['locality_name'].isna()].head()
```

Не похоже на ошибки, однако всё равно отбросим строки с этими пропусками

```python
print(raw_data['ceiling_height'].value_counts())
print(raw_data['ceiling_height'].value_counts().sort_index())
```

```python
data = raw_data.copy() # чтобы избежать изменения исходного датафрейма в процессе анализа
```

`ceiling_height` – большой массив данных, избавляться от которого неразумно, в то же время выбросов мало, так что в медиане нет особого смысла; **заменим средним арифметическим**

```python
mean_ceiling_height = data['ceiling_height'].mean()
data['ceiling_height'] = data['ceiling_height'].fillna(mean_ceiling_height)
```

`floors_total` – этих пропусков не очень много, а на факт покупки этот фактор влияет существенно; **выкинем их**

```python
data = data.dropna(subset=['floors_total'])
```

`living_area` – важный при расчётах фактор, однако почти 2000 строк; **оставим пропусками**

`is_apartment` – большое число пропусков, которые надо заменить либо строками "unknown", либо значениями False; **заменим "unknown"** (True и False тоже придётся поменять на Yes и No для гомогенности)

```python
data['is_apartment'] = data['is_apartment'].fillna('unknown')
data['is_apartment'] = data['is_apartment'].replace({True: 'Yes', False: 'No'})
```

`kitchen_area` – важный параметр, **заменим медианой**

```python
data['kitchen_area'] = data['kitchen_area'].fillna(data['kitchen_area'].median())
```

`balcony` – если балкон не указан, то скорее всего, его нет; **заменим нулями**

```python
data['balcony'] = data['balcony'].fillna(0.0)
```

`locality_name` – необходимый для статистики фактор, однако пропусков мало; **выкинем их**

```python
data = data.dropna(subset=['locality_name'])
```

`airports_nearest` – расстояние до ближайшего аэропорта не является существенным фактором; **заменим медианой** (желательно целое число)

```python
data['airports_nearest'] = data['airports_nearest'].fillna(data['airports_nearest'].median())
```

`cityCenters_nearest` – важнее аэропортов, **заменим медианами внутри населённых пунктов**

```python
# warnings.filterwarnings('ignore', category=RuntimeWarning)
data['cityCenters_nearest'] = (data
                               .groupby('locality_name')['cityCenters_nearest']
                               .transform(lambda x: x.fillna(x.median())))
```

`parks_around3000` – парки являются преимуществом, если они не указаны, значит их нет; **заменим нулями**

```python
data['parks_around3000'] = data['parks_around3000'].fillna(0.0)
```

`parks_nearest` – аналогично **заменим медианами внутри населённых пунктов**

```python
# warnings.filterwarnings('ignore', category=RuntimeWarning)
data['parks_nearest'] = (data
                         .groupby('locality_name')['parks_nearest']
                         .transform(lambda x: x.fillna(x.median())))
```

`ponds_around3000` – аналогичо паркам **заменим нулями**

```python
data['ponds_around3000'] = data['ponds_around3000'].fillna(0.0)
```

`ponds_nearest` – аналогично паркам **заменим медианами внутри населённых пунктов**

```python
# warnings.filterwarnings('ignore', category=RuntimeWarning)
data['ponds_nearest'] = (data
                         .groupby('locality_name')['ponds_nearest']
                         .transform(lambda x: x.fillna(x.median())))
```

`days_exposition` – необходимый для расчётов фактор, больше 3000 строк, **оставим пропусками**

```python
data[data['cityCenters_nearest'].isna()]
```

Видимо, для этих пропусков не дана именно локация самой квартиры.

```python
data.query('locality_name == "городской посёлок Янино-1"')
```

Возможно, точное местоположение не дают именно в небольших посёлках. Это не является проблемой для покупателей, потому что в деревнях парками и прудами редко интересуются.

- Заменить эти пропуски нулём не получится, потому что это не отражает действительности и может сказаться в процессе поиска взаимосвязей. 
- Заменить значением "unknown" – тоже не лучший вариант, потому что в других строках числа, с которыми понадобится проводить арифметические операции. 
- Выкидывать эти строки точно нельзя: пропадут данные о целых деревнях. 

Придётся оставить как есть.

Отмечу, что часть пропусков метод всё же убрал. Видимо, для крупных городов, где это было возможно.

Тем не менее для некоторых нужных параметров (таких как жилплощадь, время действия предложения или расстояние до центра города) пропуски пришлось оставить, поскольку они составляли существенную долю данных (распределнённую так же, как и основной массив).

Часть из них была заменена медианой по населённому пункту, однако сработало это не для всех случаев, потому что именно в малых городах и посёлках расстояния до центра, парков и прудов не указывает никто.

В том, что касается самих чисел, аномальных значений было не так уж много. Единичные аномалии не составило труда удалить или преобразовать в реалистичные числа (в случае высоты потолков), однако для большей части данных трудно было найти границу, до которой есть смысл рассматривать предложение, а после – нет. В итоге квартиры с более, чем 7 комнатами остались без внимания в этой работе.
